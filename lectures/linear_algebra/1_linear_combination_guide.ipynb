### Understanding Vectors: From Fundamentals to Computation

#### What Are Vectors?

At their core, vectors are mathematical objects that possess both magnitude (size) and direction. While this might seem abstract, vectors appear in many contexts in our daily lives. For instance, when you travel 5 kilometers northeast, you're essentially describing a vector - the magnitude is 5 kilometers, and the direction is northeast.

In mathematics, we typically represent vectors using arrows drawn from an origin point. The length of the arrow indicates the magnitude, while the way the arrow points shows the direction. In computational contexts, we write vectors as ordered lists of numbers called components.

For example, in two dimensions, a vector might be written as:

$$\vec{v} = (3, 4)$$

This means go 3 units in the x-direction (horizontally) and 4 units in the y-direction (vertically).

#### Vector Notation and Representation

Vectors can be denoted in several ways:
- With an arrow above the symbol: $\vec{v}$
- In bold: $\mathbf{v}$
- With underline: $\underline{v}$

When working with vectors in different dimensions:
- A 2D vector has two components: $\vec{v} = (v_1, v_2)$
- A 3D vector has three components: $\vec{v} = (v_1, v_2, v_3)$
- An n-dimensional vector has n components: $\vec{v} = (v_1, v_2, ..., v_n)$


### Basic Vector Operations

Let's explore some fundamental operations with vectors:

1. **Vector Addition**: When we add two vectors, we add their corresponding components:
   $$\vec{v} + \vec{w} = (v_1 + w_1, v_2 + w_2, ..., v_n + w_n)$$

   Geometrically, this is like following one vector and then following the other.

2. **Scalar Multiplication**: When we multiply a vector by a scalar (regular number) c, we multiply each component by c:
   $$c\vec{v} = (c \cdot v_1, c \cdot v_2, ..., c \cdot v_n)$$

   This changes the magnitude of the vector (making it longer or shorter) and potentially reverses its direction (if c is negative).

3. **Dot Product**: The dot product of two vectors gives a scalar:
   $$\vec{v} \cdot \vec{w} = v_1w_1 + v_2w_2 + ... + v_nw_n$$

   The dot product relates to the angle between vectors and has important applications in determining perpendicularity and projections.


###### Vector Spaces: A Complete Foundation

###### What is a Vector Space? The Mathematical Structure

A vector space is a mathematical structure that formalizes our intuitive notion of "things we can add and scale." More precisely, it's a set $V$ equipped with two operations that satisfy eight fundamental properties.

**The Two Operations:**
1. **Vector Addition**: For any vectors $\vec{u}, \vec{v} \in V$, their sum $\vec{u} + \vec{v}$ is also in $V$
2. **Scalar Multiplication**: For any vector $\vec{v} \in V$ and scalar $c$, the product $c\vec{v}$ is also in $V$

**The Eight Axioms:**
These operations must satisfy specific rules:
1. **Closure under addition**: $\vec{u} + \vec{v} \in V$
2. **Closure under scalar multiplication**: $c\vec{v} \in V$
3. **Associativity**: $(\vec{u} + \vec{v}) + \vec{w} = \vec{u} + (\vec{v} + \vec{w})$
4. **Commutativity**: $\vec{u} + \vec{v} = \vec{v} + \vec{u}$
5. **Zero vector exists**: There exists $\vec{0}$ such that $\vec{v} + \vec{0} = \vec{v}$
6. **Additive inverses exist**: For each $\vec{v}$, there exists $-\vec{v}$ such that $\vec{v} + (-\vec{v}) = \vec{0}$
7. **Scalar multiplication distributes**: $c(\vec{u} + \vec{v}) = c\vec{u} + c\vec{v}$
8. **Scalar addition distributes**: $(c + d)\vec{v} = c\vec{v} + d\vec{v}$

###### Concrete Examples Beyond $\mathbb{R}^n$

**Euclidean Spaces $\mathbb{R}^n$:**
- $\mathbb{R}^2$: All ordered pairs $(x, y)$ where $x, y \in \mathbb{R}$
- $\mathbb{R}^3$: All ordered triples $(x, y, z)$ where $x, y, z \in \mathbb{R}$
- Operations: $(x_1, y_1) + (x_2, y_2) = (x_1 + x_2, y_1 + y_2)$ and $c(x, y) = (cx, cy)$

**Polynomial Spaces:**
The set of all polynomials of degree at most $n$ forms a vector space:
- Vectors: $p(x) = a_0 + a_1x + a_2x^2 + ... + a_nx^n$
- Addition: $(p + q)(x) = p(x) + q(x)$
- Scalar multiplication: $(cp)(x) = c \cdot p(x)$

**Function Spaces:**
The set of all continuous functions from $\mathbb{R}$ to $\mathbb{R}$ forms an infinite-dimensional vector space:
- Vectors: Functions like $f(x) = \sin(x)$ or $g(x) = x^2$
- Addition: $(f + g)(x) = f(x) + g(x)$
- Scalar multiplication: $(cf)(x) = c \cdot f(x)$

###### Understanding Span: The Reachable Universe

**Mathematical Definition:**
The span of vectors $\{\vec{v_1}, \vec{v_2}, ..., \vec{v_k}\}$ is:
$$\text{span}\{\vec{v_1}, \vec{v_2}, ..., \vec{v_k}\} = \{c_1\vec{v_1} + c_2\vec{v_2} + ... + c_k\vec{v_k} : c_i \in \mathbb{R}\}$$

This is the set of all possible linear combinations of the given vectors.

###### Geometric Visualization of Span

**Single Vector Span:**
Consider $\vec{v} = (2, 1)$ in $\mathbb{R}^2$:
- $\text{span}\{\vec{v}\} = \{c(2, 1) : c \in \mathbb{R}\} = \{(2c, c) : c \in \mathbb{R}\}$
- This creates the line through the origin with slope $\frac{1}{2}$
- Points on this line: $(0, 0), (2, 1), (4, 2), (-2, -1), (1, 0.5)$, etc.

**Two Vector Span:**
Consider $\vec{v_1} = (1, 0)$ and $\vec{v_2} = (0, 1)$ in $\mathbb{R}^2$:
- $\text{span}\{\vec{v_1}, \vec{v_2}\} = \{c_1(1, 0) + c_2(0, 1) : c_1, c_2 \in \mathbb{R}\}$
- $= \{(c_1, c_2) : c_1, c_2 \in \mathbb{R}\} = \mathbb{R}^2$
- This spans the entire 2D plane

**Three Vector Span in 3D:**
Consider $\vec{v_1} = (1, 0, 0)$, $\vec{v_2} = (0, 1, 0)$, $\vec{v_3} = (0, 0, 1)$:
- Their span is all of $\mathbb{R}^3$
- Any point $(x, y, z) = x(1, 0, 0) + y(0, 1, 0) + z(0, 0, 1)$

###### Linear Dependence and Independence

**Key Question:** When do vectors add new "directions" to the span?

**Linear Independence:**
Vectors $\{\vec{v_1}, \vec{v_2}, ..., \vec{v_k}\}$ are linearly independent if:
$$c_1\vec{v_1} + c_2\vec{v_2} + ... + c_k\vec{v_k} = \vec{0} \text{ implies } c_1 = c_2 = ... = c_k = 0$$

**Linear Dependence:**
Vectors are linearly dependent if one can be written as a linear combination of the others.

**Example:**
- $\vec{v_1} = (1, 2)$, $\vec{v_2} = (2, 4)$ are linearly dependent because $\vec{v_2} = 2\vec{v_1}$
- $\vec{v_1} = (1, 0)$, $\vec{v_2} = (0, 1)$ are linearly independent

###### Determining if a Vector is in a Span

**The Fundamental Question:**
Given vectors $\vec{v_1}, \vec{v_2}, ..., \vec{v_k}$ and a target vector $\vec{t}$, is $\vec{t}$ in $\text{span}\{\vec{v_1}, \vec{v_2}, ..., \vec{v_k}\}$?

**Mathematical Approach:**
We need to solve: $c_1\vec{v_1} + c_2\vec{v_2} + ... + c_k\vec{v_k} = \vec{t}$

**Concrete Example:**
Is $\vec{t} = (5, 7)$ in $\text{span}\{(1, 2), (3, 1)\}$?

Set up the equation:
$$c_1(1, 2) + c_2(3, 1) = (5, 7)$$
$$c_1 + 3c_2 = 5$$
$$2c_1 + c_2 = 7$$

Solve the system:
- From first equation: $c_1 = 5 - 3c_2$
- Substitute: $2(5 - 3c_2) + c_2 = 7$
- Simplify: $10 - 6c_2 + c_2 = 7$
- Solve: $-5c_2 = -3$, so $c_2 = \frac{3}{5}$
- Therefore: $c_1 = 5 - 3(\frac{3}{5}) = 5 - \frac{9}{5} = \frac{16}{5}$

**Verification:**
$$\frac{16}{5}(1, 2) + \frac{3}{5}(3, 1) = \left(\frac{16}{5} + \frac{9}{5}, \frac{32}{5} + \frac{3}{5}\right) = (5, 7)$$

Since we found a solution, $\vec{t}$ is indeed in the span.

###### Basis and Dimension

**Basis:**
A basis for a vector space $V$ is a set of vectors that:
1. Spans $V$ (every vector in $V$ can be written as their linear combination)
2. Is linearly independent (no vector is redundant)

**Dimension:**
The dimension of a vector space is the number of vectors in any basis. All bases have the same number of vectors.

**Examples:**
- $\mathbb{R}^2$ has dimension 2; standard basis: $\{(1, 0), (0, 1)\}$
- $\mathbb{R}^3$ has dimension 3; standard basis: $\{(1, 0, 0), (0, 1, 0), (0, 0, 1)\}$
- Polynomials of degree ≤ $n$ have dimension $n+1$; basis: $\{1, x, x^2, ..., x^n\}$

###### Applications in Machine Learning and Data Science

**High-Dimensional Data:**
- Each data point is a vector in $\mathbb{R}^n$ where $n$ is the number of features
- The span of data vectors represents all possible linear combinations of your dataset
- Principal Component Analysis finds directions (basis vectors) that capture most variation

**Linear Transformations:**
- Neural network layers perform linear transformations on input vectors
- The range of a linear transformation is the span of its column vectors
- Understanding span helps determine what outputs are possible from given inputs

**Feature Engineering:**
- Adding new features increases the dimension of your vector space
- Linear dependence among features indicates redundancy
- Feature selection often involves finding a minimal spanning set

Vector spaces and span provide the mathematical foundation for understanding how linear combinations work, which is fundamental to linear algebra, machine learning, and many areas of applied mathematics. The key insight is that span describes all the "places you can reach" through linear combinations, making it a powerful tool for understanding the structure and capabilities of linear systems.

###### What Span Actually Means

**Span = All possible linear combinations**

The span of vectors is **all the places you can reach** by adding and scaling those vectors. The geometric shape depends on how many vectors you have and their relationships:

**For 2 vectors in 3D space:**
- If they're **linearly independent** (not parallel): Their span is a **plane** through the origin
- If they're **linearly dependent** (parallel): Their span is a **line** through the origin

**For 3 vectors in 3D space:**
- If all 3 are **linearly independent**: Their span is the **entire 3D space**
- If only 2 are independent: Their span is a **plane**
- If all are parallel: Their span is a **line**

###### Beyond Planes: General Span Concept

**1D span:** Single vector → Line
**2D span:** Two independent vectors → Plane  
**3D span:** Three independent vectors → Entire 3D space
**4D span:** Four independent vectors in $\mathbb{R}^4$ → Entire 4D space

###### Key Insight

Span isn't just about planes - it's about **reachability through linear combinations**. The plane is just what you get when you have exactly two independent vectors in 3D space. With different numbers of vectors or in different dimensions, span creates lines, planes, hyperplanes, or entire spaces.

So while your intuition about the plane is correct for this specific case, span is the broader mathematical concept of "everywhere you can reach by combining the given vectors."

###### Key Concept Explanations

**`A = np.column_stack(vectors)`:** This transforms a list of vectors into a matrix where each input vector becomes a column. For vectors `[1,2,1]` and `[3,1,2]`, this creates:
```
A = [[1, 3],
     [2, 1], 
     [1, 2]]
```

This matrix representation allows us to use matrix rank theory to test span membership.

**`np.all(np.abs(residuals) < 1e-10)`:** This checks if all residual values are smaller than 1e-10 (a very small number). Residuals measure how well our solution fits - small residuals mean we found coefficients that accurately reproduce the target vector. The `np.all()` ensures every component meets this precision requirement.

**What `is_in_span()` does:** Tests whether adding the target vector to our set of vectors increases the dimension of the space they span. If the dimension stays the same, the target was already reachable using the existing vectors.

**What `find_coefficients()` does:** Solves the equation "target = c₁×vector₁ + c₂×vector₂ + ..." to find the specific coefficients c₁, c₂, etc. that tell us exactly how to combine the input vectors to get the target vector.

````python
import numpy as np
from numpy.linalg import solve, lstsq, matrix_rank

def is_in_span(vectors, target):
   """
   Determine if a target vector lies within the span of a given set of vectors.
   
   This function uses the fundamental theorem of linear algebra: a vector is in the 
   span of other vectors if and only if adding it to the set doesn't increase the 
   rank (dimension) of the vector space they span.
   
   Mathematical Foundation:
       Given vectors v₁, v₂, ..., vₖ and target vector t, we ask:
       "Does there exist scalars c₁, c₂, ..., cₖ such that t = c₁v₁ + c₂v₂ + ... + cₖvₖ?"
       
       This is equivalent to asking if the system Ax = t has a solution, where:
       - A is the matrix with vectors as columns: A = [v₁ | v₂ | ... | vₖ]
       - x is the vector of coefficients [c₁, c₂, ..., cₖ]
       
       By the rank-nullity theorem, this system has a solution if and only if
       rank(A) = rank([A | t]), where [A | t] is the augmented matrix.
   
   Parameters:
       vectors (list of numpy.ndarray): List of vectors that define the span.
           Each vector should be a 1D numpy array of the same length.
           Example: [np.array([1, 2]), np.array([3, 4])]
           
       target (numpy.ndarray): The vector to test for membership in the span.
           Must have the same dimension as the vectors in the vectors list.
           Example: np.array([5, 6])
   
   Returns:
       bool: True if target vector is in the span of the given vectors,
             False otherwise.
   
   Raises:
       ValueError: If vectors have different dimensions or if target dimension
                  doesn't match the vector dimensions.
       
   Algorithm Details:
       1. Stack input vectors as columns to form matrix A
       2. Create augmented matrix [A | target]
       3. Compare rank(A) with rank([A | target])
       4. If ranks are equal, target is in span; otherwise it's not
       
   Time Complexity: O(min(m,n)³) where m is number of rows, n is number of columns
   Space Complexity: O(mn) for storing matrices
   
   Example:
       >>> v1 = np.array([1, 2])
       >>> v2 = np.array([3, 4])
       >>> target = np.array([7, 10])  # This is 1*v1 + 2*v2
       >>> is_in_span([v1, v2], target)
       True
       
       >>> target2 = np.array([1, 0])  # Cannot be written as combination of v1, v2
       >>> is_in_span([v1, v2], target2)  
       False
   """
   # Convert list of vectors to matrix where each vector becomes a column
   # This transforms the span question into a matrix rank comparison
   A = np.column_stack(vectors)
   
   # Create augmented matrix [A | target] by adding target as the last column
   # The augmented matrix represents the system Ax = target
   augmented = np.column_stack((A, target))
   
   # Core theorem: target is in span(vectors) ⟺ rank(A) = rank([A | target])
   # If adding target increases the rank, then target adds a new dimension
   # and is therefore not expressible as a linear combination of existing vectors
   return matrix_rank(A) == matrix_rank(augmented)


def find_coefficients(vectors, target):
   """
   Find scalar coefficients to express target vector as linear combination of given vectors.
   
   This function solves the linear system Ax = target where A contains the input vectors
   as columns and x contains the coefficients we want to find. The solution x tells us
   how to combine the input vectors to reconstruct the target vector.
   
   Mathematical Foundation:
       We seek coefficients c₁, c₂, ..., cₖ such that:
       target = c₁ × vector₁ + c₂ × vector₂ + ... + cₖ × vectorₖ
       
       In matrix form: A × c = target, where:
       - A = [vector₁ | vector₂ | ... | vectorₖ] (each vector is a column)
       - c = [c₁, c₂, ..., cₖ]ᵀ (coefficients we want to find)
       - target = given target vector
       
       We use least squares method which finds the best solution even when
       the system is overdetermined (more equations than unknowns) or when
       no exact solution exists due to numerical precision issues.
   
   Parameters:
       vectors (list of numpy.ndarray): List of basis vectors for the linear combination.
           Each vector should be a 1D numpy array of the same length.
           The number of vectors determines how many coefficients will be returned.
           
       target (numpy.ndarray): The target vector to express as a linear combination.
           Must have the same dimension as each vector in the vectors list.
   
   Returns:
       numpy.ndarray or None: 
           - If a solution exists: Array of coefficients [c₁, c₂, ..., cₖ] such that
             c₁×vector₁ + c₂×vector₂ + ... + cₖ×vectorₖ ≈ target
           - If no solution exists: None
           
       The returned coefficients can be used to reconstruct the target:
       reconstruction = sum(coeff * vector for coeff, vector in zip(coefficients, vectors))
   
   Algorithm Details:
       1. Stack vectors as columns to form coefficient matrix A
       2. Use NumPy's lstsq (least squares) to solve Ax = target
       3. Check if residuals are negligible (< 1e-10) to determine if exact solution exists
       4. Return coefficients if solution found, None otherwise
       
   Numerical Considerations:
       - Uses least squares instead of direct solve() to handle edge cases gracefully
       - Tolerance of 1e-10 accounts for floating-point precision limitations
       - Works even when system is slightly overdetermined due to numerical noise
       
   Time Complexity: O(min(m,n)³) for SVD-based least squares solution
   Space Complexity: O(mn) for matrix storage
   
   Example:
       >>> v1 = np.array([1, 0])
       >>> v2 = np.array([0, 1])  
       >>> target = np.array([3, 4])
       >>> coeffs = find_coefficients([v1, v2], target)
       >>> print(f"target = {coeffs[0]} * v1 + {coeffs[1]} * v2")
       target = 3.0 * v1 + 4.0 * v2
       
       >>> # Verify reconstruction
       >>> reconstruction = coeffs[0] * v1 + coeffs[1] * v2
       >>> np.allclose(reconstruction, target)
       True
   """
   # Convert input vectors to coefficient matrix A where each vector is a column
   # This sets up the matrix equation Ax = target where x are the coefficients we seek
   A = np.column_stack(vectors)
   
   # Use least squares to solve Ax = target
   # lstsq returns: (solution, residuals, rank, singular_values)
   # - solution: the coefficients that best satisfy the equation
   # - residuals: sum of squared residuals (measure of how well solution fits)
   # - rank: effective rank of matrix A
   # - singular_values: singular values from SVD decomposition
   coefficients, residuals, rank, singular_values = lstsq(A, target, rcond=None)
   
   # Check solution quality using residuals
   # residuals array contains sum of squared residuals for each target column
   # Empty residuals array means the system is exactly determined (m <= n)
   # Small residuals (< 1e-10) indicate we found a good approximate solution
   #
   # np.abs(residuals) < 1e-10: Element-wise comparison creating boolean array
   # np.all(...): Returns True only if ALL elements satisfy the condition
   # This ensures every component of the solution has acceptable error
   if len(residuals) == 0 or np.all(np.abs(residuals) < 1e-10):
       return coefficients
   
   # If residuals are too large, no acceptable solution exists
   # This happens when target is not in the span of the given vectors
   return None


# Demonstration and verification
if __name__ == "__main__":
   # Define test vectors in R³
   v = np.array([1, 2, 1])  # First basis vector
   w = np.array([3, 1, 2])  # Second basis vector  
   t = np.array([5, 7, 5])  # Target vector to test
   
   # Test span membership
   print(f"Is t in the span of v and w? {is_in_span([v, w], t)}")
   
   # Find explicit linear combination coefficients
   coeffs = find_coefficients([v, w], t)
   if coeffs is not None:
       print(f"t = {coeffs[0]:.3f} × v + {coeffs[1]:.3f} × w")
       
       # Verify our solution by reconstruction
       reconstruction = coeffs[0] * v + coeffs[1] * w
       print(f"Original target:     {t}")
       print(f"Reconstructed:       {reconstruction}")
       print(f"Reconstruction error: {np.linalg.norm(t - reconstruction):.2e}")
   else:
       print("t is not in the span of v and w")

# Example usage
v = np.array([1, 2, 1])
w = np.array([3, 1, 2])
t = np.array([5, 7, 5])

print(f"Is t in the span of v and w? {is_in_span([v, w], t)}")

````

#### Checking if A Set Forms a Vector Space

Using Python, we can verify whether a given set satisfies the vector space properties:

````python
import numpy as np
from numpy import linalg as LA

def is_closed_under_addition(vectors):
    """Check if the set is closed under addition."""
    for v in vectors:
        for w in vectors:
            v_plus_w = v + w
            if not any(np.array_equal(v_plus_w, x) for x in vectors):
                return False
    return True

def is_closed_under_scalar_multiplication(vectors, scalars):
    """Check if the set is closed under scalar multiplication."""
    for v in vectors:
        for s in scalars:
            s_times_v = s * v
            if not any(np.array_equal(s_times_v, x) for x in vectors):
                return False
    return True

# Example: Check if the set of standard basis vectors in ℝ² forms a vector space
e1 = np.array([1, 0])
e2 = np.array([0, 1])
vectors = [e1, e2]

# This is NOT a vector space because it's not closed under addition
# e1 + e2 = [1, 1] which is not in our set
print(f"Closed under addition? {is_closed_under_addition(vectors)}")

# Let's try a more comprehensive set
# The entire ℝ² is too large to represent explicitly, but we can create a finite subset
def generate_integer_vectors_in_range(n):
    """Generate all vectors with integer coordinates in [-n, n]."""
    result = []
    for i in range(-n, n+1):
        for j in range(-n, n+1):
            result.append(np.array([i, j]))
    return result

# Create a subset of ℝ² with integer coordinates in [-2, 2]
integer_vectors = generate_integer_vectors_in_range(2)
print(f"Number of vectors in our subset: {len(integer_vectors)}")

# Check if this subset satisfies closure under addition within the subset
print(f"Is our subset closed under addition? {is_closed_under_addition(integer_vectors)}")

# It's not, because e.g. [2,2] + [1,1] = [3,3] which is outside our range
# This illustrates that we need the entire ℝ² to have a proper vector space
````

#### Understanding Vector Spans in Linear Algebra

The concept of span represents one of the foundational pillars of linear algebra, with profound implications across mathematical theory and applications. Let me develop this concept systematically, beginning with its formal definition and progressing to its geometric interpretation.

##### Formal Definition

In linear algebra, the span of a set of vectors is the set of all possible linear combinations of those vectors. More precisely, given vectors $\vec{v}_1, \vec{v}_2, ..., \vec{v}_n$ in a vector space $V$, the span is defined as:

$$\text{span}(\vec{v}_1, \vec{v}_2, ..., \vec{v}_n) = \{c_1\vec{v}_1 + c_2\vec{v}_2 + ... + c_n\vec{v}_n \mid c_1, c_2, ..., c_n \in \mathbb{R}\}$$

This mathematical formulation tells us that the span consists of all vectors that can be expressed as linear combinations of the original vectors, where the coefficients $c_1, c_2, ..., c_n$ can be any real numbers.

##### Geometric Interpretation

The geometric interpretation of span provides intuitive understanding:

1. The span of a single nonzero vector $\vec{v}$ in $\mathbb{R}^n$ is a line passing through the origin in the direction of $\vec{v}$.

2. The span of two linearly independent vectors in $\mathbb{R}^n$ forms a plane passing through the origin.

3. The span of three linearly independent vectors in $\mathbb{R}^3$ constitutes the entire three-dimensional space.

This progression illustrates how spans create subspaces of increasing dimension.

##### Determining if a Vector Lies Within a Span

In your specific question, determining whether vector $\vec{t}$ lies within the span of vectors $\vec{v}$ and $\vec{w}$ requires ascertaining if there exist scalars $a$ and $b$ such that:

$$\vec{t} = a\vec{v} + b\vec{w}$$

This is equivalent to solving the system of linear equations that arises when we decompose this vector equation into its component equations. If a solution exists, then $\vec{t}$ lies within the span; if no solution exists, then $\vec{t}$ is outside the span.

##### Analytical Example

Consider vectors $\vec{v} = (1,2,1)$, $\vec{w} = (3,1,2)$, and we wish to determine if $\vec{t} = (5,7,5)$ lies within their span.

We need to find if there exist values $a$ and $b$ such that:
$$a(1,2,1) + b(3,1,2) = (5,7,5)$$

This gives us the system:
$$a + 3b = 5$$
$$2a + b = 7$$
$$a + 2b = 5$$

Solving this system (using substitution or elimination) would determine whether $\vec{t}$ lies within the span of $\vec{v}$ and $\vec{w}$.

##### Theoretical Significance

The concept of span is intrinsically connected to other fundamental concepts in linear algebra:

1. A set of vectors is linearly independent if none of the vectors can be expressed as a linear combination of the others.

2. The dimension of a vector space is the cardinality of its smallest spanning set.

3. A spanning set for a vector space $V$ is a set of vectors whose span is the entire space $V$.

These relationships form the backbone of linear algebra's theoretical structure and find applications in fields ranging from computer graphics to quantum mechanics.

````python
def check_vector_span(set_of_vectors, vector_to_check):
    # Creates an empty vector of correct size
    vector_of_scalars = np.asarray([None]*set_of_vectors.shape[0])
    
    # Solves for the scalars that make the equation true if vector is within the span
    try:
        # Use np.linalg.solve() function here to solve for vector_of_scalars
        vector_of_scalars = np.linalg.solve(set_of_vectors, vector_to_check)
        if not (vector_of_scalars is None):
            print("\nVector is within span.\nScalars in s:", vector_of_scalars)
        
    # Handles the cases when the vector is NOT within the span   
    except Exception as exception_type:
        if str(exception_type) == "Singular matrix":
            print("\nNo single solution\nVector is NOT within span")
        else:
            print("\nUnexpected Exception Error:", exception_type)
    return vector_of_scalars
````

## System of Equations
All the cases that we tested above could have also been written as a system of two equations, where we are trying to solve for the values of the scalars that make both equations true. For the system of equations, scalar $a$ becomes $x$ and scalar $b$ becomes $y$.

So *Equation 1*, $a\vec{v} + b\vec{w} = \vec{t}$, which could be written as: 
    
$\hspace{1cm} a \begin{bmatrix} 1\\ 3\end{bmatrix} + b \begin{bmatrix} 2\\ 5\end{bmatrix} = \begin{bmatrix} 4\\ 11\end{bmatrix}$, where $a = 2$ and $b = 1$ 

Becomes the following system of two equations that is written as:

$\hspace{1cm} \begin{array}{rcl} x + 2y & = & 4 \\ 3x + 5y  & = & 11 \end{array}$, where $x = 2$ and $y = 1$ 

*__Notice that__*:

- The vectors $\vec{v}$ and $\vec{w}$ become the coefficients on the *left* side of both equations. 

- The vector $\vec{t}$ become the solution on the *right* side of both equations. 

- The scalar $a$ becomes the variable $x$ and the scalar $b$ becomes variable $y$ in both equations.

- Each of the equations can be represented by a line plotted in two dimensions.


Systems of equations always result in *one* of *three* possible solutions. One occurs when the vector is within the span and there's a solution, like with the example above. The other two cases can occur when the vector is **not** within span. Below we describe each of the three cases.


### Case 1 - Single solution
We could have considered *Equation 1* as the following system of two equations:

$\hspace{1cm} \begin{array}{rcl} x + 2y & = & 4 \\ 3x + 5y  & = & 11 \end{array}$, where $x = 2$ and $y = 1$ 

We would have used the same method to solve this system of equations for $x$ and $y$, as we did to determine vector $\vec{t}$'s span. This means when the vector is within the span, there is a single solution to the system of equations. This single solution graphically is represented where the lines intersect (the red dot on the graph below). 

````python
%matplotlib inline
import matplotlib.pyplot as plt

plt.plot([4,0],[0,2],'b',linewidth=3)
plt.plot([3.6667,0],[0,2.2],'c-.',linewidth=3)

plt.plot([2],[1],'ro',linewidth=3)
plt.xlabel('Single Solution')
plt.show()
````

### Case 2 - Infinite Solutions
The second case is when there are infinite values that the scalars could have taken because at least two of the equations are redundant. In our case, our only two equations are redundant and they represent the same line (see graph below). 

This second case is represented by $vw2$ and $t2$ where:
    
$\hspace{1cm} \begin{array}{rcl} x + 2y & = & 6 \\ 2x + 4y  & = & 12 \end{array}$, where **infinite** $x$ and $y$ makes this *__true__* because the equations are redundant. 

````python
import matplotlib.pyplot as plt

plt.plot([6,0],[0,3],'b',linewidth=5)
plt.plot([1,4,6,0],[2.5,1,0,3],'c-.',linewidth=2)
plt.xlabel('Redundant Equations')
plt.show()
````

### Case 3 - No Solution
The third case is that there are **no** values that the scalars could have taken that would have simutaneously solved all equations.
In our case, our only two equations are represented by parallel lines because they have no solution (see graph below). 

This third case is represented by $vw3$ and $t3$ where:
    
$\hspace{1cm} \begin{array}{rcl} x + 2y & = & 6 \\ x + 2y  & = & 10 \end{array}$, where **no** $x$ and $y$ make this true. 

````python
import matplotlib.pyplot as plt

plt.plot([10,0],[0,5],'b',linewidth=3)
plt.plot([0,6],[3,0],'c-.',linewidth=3)
plt.xlabel('No Solution')
plt.show()
````
